##### Indexing Slicing Transformation

```
# Choose one value in a matrix
arr_bis = np.array([[1,2,3],
                    [7,1,6],
                    [9,6,3]])
arr_bis[0][0]
```

```
arr_bis[0,0]
```

```
# Choose sub matrix in the matrix
arr_bis[0:2,0:1]
```

```
# Choose one columns or one row
print(arr_bis[1,:])
print(arr_bis[:,1])
```

```
# Max
# On the matrix
print(arr_bis.max())

# By the rows
print(arr_bis.max(axis=1))

# By the columns
print(arr_bis.max(axis=0))
```

```
# Min
# On the matrix
print(arr_bis.min())

# By the rows
print(arr_bis.min(axis=1))

# By the columns
print(arr_bis.min(axis=0))
```

```
# Mean
# On the matrix
print(arr_bis.mean())

# By the rows
print(arr_bis.mean(axis=1))

# By the columns
print(arr_bis.mean(axis=0))
```

```
# Std
# On the matrix
print(arr_bis.std())

# By the rows
print(arr_bis.std(axis=1))

# By the columns
print(arr_bis.std(axis=0))
```

```
# Other way available 
np.std(arr_bis, axis=0)
```

```
# Log function 
np.log(arr_bis)
```

```
# Exponential function 
np.exp(arr_bis)
```

```
# Squared root function 
np.sqrt(arr_bis)
```

```
# Concatenate 
# Params: tuple of arrays and the axis of the concatenation
# Concat 1 dimension
arr1 = arr_bis[:,1]
arr2 = arr_bis[:,2]

print(f"ARR1 {arr1}")
print(f"ARR2 {arr2}")

print(np.concatenate((arr1, arr2), axis=0))
```

```
# Concat 2 dimension
arr1 = arr1.reshape(-1,1)
arr2 = arr2.reshape(-1,1)

print(f"ARR1 {arr1}")
print(f"ARR2 {arr2}")

print(np.concatenate((arr1, arr2), axis=0))
```

```
# Concat 2 dimension
print(np.concatenate((arr1, arr2), axis=1))
```

##### Cleaning and selection

```
# Import of csv
assets = pd.read_csv("/content/assets.csv", parse_dates=True, index_col="time")
assets
```

```
# Ordonning
assets = assets.sort_index(ascending=True)
assets
```

```
# Select a column
assets = assets[["Close DJI30", "Close CAC40", "Close SP500"]]
assets
```

```
# Missing values NAN
assets = assets.dropna()
assets
```

```
# Reset index
assets.reset_index(drop=True)
```

```
# Rolling 
assets["SMA 15"] = assets[["Close SP500"]].rolling(15).mean()
assets
```

```
# Shift
assets["Close SP500"].shift(1)
```

```
# Groupby by mean
arr = [[1,11],
       [1,10],
       [3,15],
       [3,15],
       [3,10]]

df = pd.DataFrame(arr, columns=["num","value"])

df.groupby(by="num").mean()
```

```
# Groupby by sum
df.groupby(by="num").sum()
```

```
# Groupby by std
df.groupby(by="num").std()
```

##### Iloc & Loc

```
# Slicing
assets.iloc[0:1500, 0:10]
```

```
# Conditonal by dates
assets.loc["2010":"2012"]
```

```
# Conditonal by values
assets.loc[assets["Close DJI30"] > 15000]
```

```
# Concat some dataframes
df1 = assets.iloc[:,:3]
df2 = assets.iloc[:,3:]

pd.concat((df1,df2), axis=1)
```

```
# Concat some dataframes
df1 = assets.iloc[1500:,:]
df2 = assets.iloc[:1500,:]

pd.concat((df1,df2), axis=0)
```

##### Graphs

```
# data simulation
arr = np.random.randint(0,15,size=(15,))

# square of arr
square = arr**2

# Mutltiplication of arr
mut = arr*3

print(arr)
print(square)
print(mut)
```

```
# Plot the data
plt.plot(arr)
plt.plot(square)
plt.plot(mut)

# Change the xlabel 
plt.xlabel("Random numbers")

# Change the ylabel 
plt.ylabel("Values")

# Change the title
plt.title("Graph")

# Put a legend
plt.legend(["Normal", "Square", "Multiplication"])

# Show the chart
plt.show()
```

##### Scatter

```
# Point clouds
plt.scatter(arr, mut)
plt.scatter(arr, square)


# Change the xlabel 
plt.xlabel("Random Numbers")

# Change the ylabel 
plt.ylabel("Values")

# Change the title 
plt.title("Scatter")

# Put a legend
plt.legend(["Mut", "Square"])

# Show the plot
plt.show()
```

##### Tools 

```
# Adapt the size
plt.figure(figsize=(15,8))

# Plot the line
plt.plot(arr, alpha=0.6, color="turquoise")

# Plot the line
plt.plot(square, "-o", linewidth=3)

# Plot the line
plt.plot(mut, "o")

# Change the ylabel 
plt.ylabel("Values")

# Change the title 
plt.title("Graph")

# Put a legend
plt.legend(["Normal", "Square", "Multiplication"])

# show the plot
plt.show()
```
